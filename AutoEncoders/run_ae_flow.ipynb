```python
models = [
    C2D_AE_3x3_Res(channels = 6)
         ]

model_names = [
    "C2D_AE_Res_128_3x3_FLOW",
]

model_names = [join_paths(["C2D_AE_models/", mn+"_MSE_SS"]) for mn in model_names]    
```

# Training

```python
optimizers = [torch.optim.Adam(m.parameters(), lr = 1e-3) for m in models]
losses = [nn.MSELoss() for m in models]
```

```python
train_data = StreetScene(image_type="optical_flow", frame_strides=[2,4,8], sample_stride=4)
```

```python
train_loader, val_loader = get_data_loaders(train_data, batch_size = 32)
```

```python
trainer = AutoEncoder_Trainer(
    models,
    model_names,
    train_loader, 
    val_loader,
    optimizers,
    losses,
    epochs = 300
)
```

```python
trainer.train()
```

```python
save_model(trainer.autoencoder_models[0].model, trainer.autoencoder_models[0].model_file)
```

```python
print("Training done")
```
