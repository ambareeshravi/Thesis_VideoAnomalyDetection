```python
models = [
    C2D_AE_128_3x3(channels = 1, filters_count=[64, 64, 128, 256, 128, 128]),
    C2D_AE_128_5x5(channels = 1, filters_count=[64, 64, 128, 256, 128, 128]),
         ]

model_names = [
    "C2D_3x3_2xE",
    "C2D_3x3_2xE"
]

model_names = [join_paths(["C2D_AE_models/128/L2_Regularization/", mn+"_WEIGHTED_UCSD2"]) for mn in model_names]    
```

# Training

```python
optimizers = [ADAM(m) for m in models]
losses = [WEIGHTED_SIMILARITY() for m in models]
```

```python
train_data = UCSD(2)
```

```python
train_loader, val_loader = get_data_loaders(train_data, batch_size=64)
```

```python
trainer = AutoEncoder_Trainer(
    models,
    model_names,
    train_loader, 
    val_loader,
    optimizers,
    losses,
    epochs = 200
)
```

```python
trainer.train()
```

```python
for ae_model in trainer.autoencoder_models:
    save_model(ae_model.model, ae_model.model_file)
```

```python
print("Training done")
```
