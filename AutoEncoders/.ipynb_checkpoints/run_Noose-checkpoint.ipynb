```python
import sys
sys.path.append("..")
from general.data import *
from train_autoencoder import AutoEncoder_Trainer
from test_autoencoder import AutoEncoder_Tester
torch.autograd.set_detect_anomaly(True)

from C2D_Models import *
```

```python
class NooseLoss:
    def __init__(self):
        self.mse = nn.MSELoss()
    
    def noose(self, x):
        return torch.norm(x - torch.mean(x))
    
    def __call__(self, original, reconstructions, encodings):
        return self.mse(original, reconstructions) + self.noose(encodings)
```

```python
models = [
    C2D_AE_128_3x3(channels = 1),
]

model_files = [
    "C2D_AE_3x3_Noose",
]

model_files = [join_paths(["C2D_AE_models/128/experiments/noose/", m + "_UCSD2"]) for m in model_files]
optimizers = [ADAM(m) for m in models]
losses = [NooseLoss() for m in models]
```

```python
ucsd = UCSD(2, sample_stride=1)
train_loader, val_loader = get_data_loaders(ucsd, 256)
```

```python
def step(images):
    reconstructions, encodings = trainer.autoencoder_models[0].model(trainer.autoencoder_models[0].get_inputs(images))
    return trainer.autoencoder_models[0].loss_criterion(images, reconstructions, encodings)
```

```python
trainer = AutoEncoder_Trainer(
    models,
    model_files,
    train_loader,
    val_loader,
    optimizers,
    losses,
    epochs = 300,
    useGPU=True
)
```

```python
trainer.autoencoder_models[0].step = step
```

```python
trainer.train()
```

# Testing Noose

```python
test_data = UCSD(2, isTrain=False, sample_stride = 1)
```

```python
for ae_model in trainer.autoencoder_models:
    print(ae_model.model_file)
    tester = AutoEncoder_Tester(
        ae_model.model,
        test_data
    )
    score = tester.test(save_as = ".pkl".join(ae_model.model_file.split(".pth.tar")))
#     print(score)
    print("-"*40)
```
