```python
from train_autoencoder import *
```

```python
from C2D_Models import C2D_AE_128_3x3_VAE
```

```python
model = C2D_AE_128_3x3_VAE(channels=3)
```

# Training

```python
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
loss_criterion = None
```

```python
train_data = Avenue(sample_stride=1)
```

```python
train_loader, val_loader = get_data_loaders(train_data)
```

```python
trainer = AutoEncoder_Trainer(
    [model],
    ["C2D_AE_models/C2D_VAE_128_3x3_AVENUE_300"],
    train_loader, 
    val_loader,
    optimizer,
    loss_criterion,
    epochs = 300
)
```

```python
trainer.train()
```

```python
print("Training done")
```
