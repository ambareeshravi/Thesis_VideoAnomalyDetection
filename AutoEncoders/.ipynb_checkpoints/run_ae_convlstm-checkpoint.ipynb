```python
models = [
    CLSTM_AE(image_size=128, channels = 1),
         ]

model_files = [
    "CLSTM_AE_L2",
]

model_files = [join_paths(["../models/ConvLSTM_models/128/", m +"_UCSD2"]) for m in model_files]
```

```python
load_model(models[0], '../models/ConvLSTM_models/128/CLSTM_AE_L2_UCSD2/CLSTM_AE_L2_UCSD2.pth.tar')
```

# Training

```python
optimizers = [ADAM(m, lr = 0.0005) for m in models]
losses = [MSE_LOSS() for m in models]
```

```python
train_data = UCSD(2, image_size = 128, asImages=False, n_frames=16)
```

```python
train_loader, val_loader = get_data_loaders(train_data, batch_size=8)
```

```python
trainer = AutoEncoder_Trainer(
    models,
    model_files,
    train_loader, 
    val_loader,
    optimizers,
    losses,
    epochs = 120,
    lr_scheduler_params = {"factor": 0.75, "patience": 3, "threshold": 1e-4},
)
```

```python
trainer.train()
```

```python
for ae_model in trainer.autoencoder_models:
    save_model(ae_model.model, ae_model.model_file)
```

```python
print("Training Done")
```
